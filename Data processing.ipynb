```python
import pandas as pd
import numpy as np
import unidecode
```

```python
df = pd.read_excel('/Users/shizhewei/Desktop/Assignments/Data Eng Final Project/Global_Superstore.xlsx')
df.head(1)
```

```python
order_date_int = []
for i in df['Order Date']:
    day = i[0:2]
    month = i[3:5]
    year = i[6:10]
    order_date_int.append(int(year + month + day))

ship_date_int = []
for i in df['Ship Date']:
    day = i[0:2]
    month = i[3:5]
    year = i[6:10]
    ship_date_int.append(int(year + month + day))

df['Order Date'] = order_date_int
df['Ship Date'] = ship_date_int

new_code = []
for i in df['Postal Code']:
    if i > 0:
        new_code.append(int(i))
    else:
        new_code.append(0)
df['Postal Code'] = new_code

df.head(3)
```

```python
# column with issues: customer_name, city, state, coountry, product_name
new_customer_name = []
for i in df['Customer Name']:
    new_customer_name.append(unidecode.unidecode(i))
df['Customer Name'] = new_customer_name
```

```python
new_city = []
for i in df['City']:
    new_city.append(unidecode.unidecode(i))
df['City'] = new_city
```

```python
new_state = []
for i in df['State']:
    new_state.append(unidecode.unidecode(i))
df['State'] = new_state
```

```python
new_country = []
for i in df['Country']:
    new_country.append(unidecode.unidecode(i))
df['Country'] = new_country
```

```python
new_prod_name = []
for i in df['Product Name']:
    new_prod_name.append(unidecode.unidecode(i))
df['Product Name'] = new_prod_name
```
